#### 7*. Согласно современным представлениям, форма Земли близка к сплюснутому сфероиду с полуосями $a = 6 378 137.0$ м и $c = 6 356 752.314245$ м соответственно. Используя формулу для площади сплюснутого сфероида $S_{obl}=2 \pi a^2(1+\frac{1-e^2}e \mathrm{atanh}(e))$, где $e^2=1-\frac{c^2}{a^2}$ вычислите площадь поверхности Земли и выясните на сколько она отличается от площади сферы радиусом 6371 км. Для вычисления гиперболического арктангенса можно использовать следующий фрагмент кода:

Для сложных задач и хадач со звёздочкой будет всегда много комментариев. Их лучше удалять а сложные переменные просто менять на однобуквенные. Я так пишу чтобы всем было понятно, что делать 

```python
import math # импортирую библиотеку для более удобной работы с математическими операторами (хештег # нужен чтоби писать комментарии)

# Заданные параметры
a = 6378137.0  # большая полуось в метрах
c = 6356752.314245  # малая полуось в метрах

# Вычисление эксцентриситета просто по формуле
e_squared = 1 - (c**2 / a**2) 
e = math.sqrt(e_squared) #sqrt это оператор квадратного корня из библиотеке math, вместо него можно просто писать (e_squared)**0.5, но это нагромаждает код, поэтому я так не делаю

# Вычисление площади поверхности эллипсоида по формуле
S_ellipsoid = 2 * math.pi * a**2 * (1 + ((1 - e_squared) / e) * math.atanh(e)) #просто формула из условия

# Вычисление площади сферы радиусом 6371 км
R_sphere = 6371000  # радиус в метрах
S_sphere = 4 * math.pi * R_sphere**2 #и площадь в квадратных метрах

# Разница между площадями
difference = S_ellipsoid - S_sphere
relative_difference = (difference / S_sphere) * 100

# Вывод результатов
print("Результаты вычислений:")
print(f"Эксцентриситет e = {e:.10f}")
print(f"e² = {e_squared:.10f}")
print(f"Площадь поверхности эллипсоида: {S_ellipsoid:.2f} м²")
print(f"Площадь поверхности сферы (R=6371 км): {S_sphere:.2f} м²")
print(f"Разница: {difference:.2f} м²")
print(f"Относительная разница: {relative_difference:.6f}%")

# Дополнительно: вывод в квадратных километрах
S_ellipsoid_km2 = S_ellipsoid / 1e6
S_sphere_km2 = S_sphere / 1e6
print(f"\nВ квадратных километрах:")
print(f"Площадь эллипсоида: {S_ellipsoid_km2:.2f} км²")
print(f"Площадь сферы: {S_sphere_km2:.2f} км²")


# Блок вывода естественно можно упростить как вам хочется, главное чтобы оно удволетворядо ТЗ от Фирсова
```
