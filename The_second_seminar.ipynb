#### 2. Температура плавления брома равна $265.9\:\mathrm{K}$, температура кипения — $332.4\:\mathrm{K}$. Напишите программу, которая запрашивает у пользователя температуру (в $\mathrm{^\circ C}$) и выводит агрегатное состояние брома.

```python
t1 = 265.9
t2 = 332.4
tu = float(input('Введите температуру вещества в градусах по цельсию')) + 273.15
if tu < t1:
    print ('Твёрдое вещество')
elif tu > t2:
    print ('Газ')
else:
    print ('Жидкость')
```

#### 3. Напишите программу, которая запрашивает у пользователя длины трёх сторон треугольника и выводит его площадь или сообщение `Не существует`, если треугольника с такими длинами сторон не существует.

```python
print ('введите длины сторон')
a = float(input('a='))
b = float(input('b='))
c = float(input('c='))
if a + b > c  and b + c > a and c + a > b :
    print ('Площадь треугольника')
    p = (a + b + c)/2
    S = (p*(p-a)*(p-b)*(p-c))**0.5
    print (S)
else:
    print ('треугольника не существует')
```

#### 4. Даны две сферы радиусом $r_1$ и $r_2$, центры сфер расположены на расстоянии $d$. Напишите программу, которая запрашивает $r_1$, $r_2$ и $d$ и выводит объём области пересечения сфер для любого случая их взаимного положения.

Задача сложная, поэтому с комментами

```python
import math
print ('введите радиусы сфер')
r1 = float(input('r1='))
r2 = float(input('r2='))
d = float(input('d='))
if d >= r1+r2: # проверяем вообще наличие пересечения у сфер, если его нет выводи 0, именно ноль, как в ТЗ у Фирсова, он любит до этого докапаться
    print (0)
elif d < r1 + r2 and d > abs(r1-r2): # abs это функция модуля, тогда мы проверяем условие, что сферы пересекаются, но не заходят друг в друга и применяем формулу из википедии 
    print(math.pi/(12*d)*(r1+r2-d)**2 * ((d**2 + 2 * d * (r1+r2) - 3*(r1-r2)**2)))
elif r1>r2 and d <= r1-r2: #отдельные случаи когда одно сфера внутри другой
    print (4/3*math.pi*r2**3)
elif r1>r2 and d <= r2+r1: #тот же случай, но учитываем вариант если размеры сфер инвертированны (тут просто с модулем не получится прописать)
    print (4/3*math.pi*r1**3)
```
