#### 5*. Создайте вектор `X`, содержащий 10 значений, равномерно распределённых на отрезке $(0,1)$, не включающий концы отрезка (★☆☆)
<center><img src="https://lk.challenges2024.ru/visualization/np_prac06.png" width=25%></center>

```python
X = np.linspace(0, 1, 12)[1:-1]
print(X)
```

#### У элементов вектора `X`, имеющих значения от $0.3$ до $0.7$, поменяйте знак на противоположный. (★☆☆)

```python
X[(0.3 < X) & (X < 0.7)] *= -1
print(X)
```

#### Создайте вектор `X`, содержащий 10 случайных значений из интервала $(0,1)$ и измените максимальное значение на $0$. (★☆☆)

```python
X = np.random.rand(10)
print (X)
X[X.argmax()] = 0
print(X)
```

#### Выведите значение вектора `X`, ближайшее к $0.5$. (★☆☆)

```python
print(X[np.abs(X - 0.5).argmin()])
```

#### Распечатайте диапазон значений стандартных числовых типов данных в Numpy. (★☆☆)

```python
for dtype in [np.int8, np.int32, np.int64]:
    print(np.iinfo(dtype).min)
    print(np.iinfo(dtype).max)
for dtype in [np.float32, np.float64]:
    print(np.finfo(dtype).min)
    print(np.finfo(dtype).max)
    print(np.finfo(dtype).eps)
```
