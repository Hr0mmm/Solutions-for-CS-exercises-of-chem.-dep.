#### <p style='text-align: justify;'>3. Напишите программу, которая запрашивает объём $V_a\:\mathrm{мл}$ и концентрацию $C_a\:\mathrm{моль/л}$ раствора $\mathrm{HCl}$ и объём $V_b\:\mathrm{мл}$ и концентрацию $C_b\:\mathrm{моль/л}$ раствора $\mathrm{NaOH}$ и рассчитывает $\mathrm{pH}$ после смешения этих растворов. Формулы для расчёта и комментарии см. в файле `share/ph.pdf`. Выведите рассчитанное значение $\mathrm{pH}$ и является ли среда кислой, щелочной или нейтральной ($\mathrm{pH}\:\:6.5-7.5$). Для вычисления десятичного логарифма можно использовать функцию `log10` из библиотеки `math`.</p>

На самом деле задачка очень простая, вся необходимая методика описана у Фирсова в файле, не поленитесь и просто прочитайте

Контрольный пример
Ca = 0.12 моль/л, Cb = 0.15 моль/л
10 мл HCl + 7.5 мл NaOH; pH = 2.368
10 мл HCl + 8.0 мл NaOH; pH = 6.998
10 мл HCl + 8.5 мл NaOH; pH = 11.604

```python
import math
Va=float(input('объём HCl = ?'))/1000
Ca=float(input('концентрация HCl = ?'))
Vb=float(input('объём NaOH = ?'))/1000
Cb=float(input('концентрация NaOH = ?'))
A=Va*Ca
B=Vb*Cb
V=Va+Vb
w=1.008e-14
p=(A-B)/V
C=abs(p/2)+math.sqrt((p**2)/4+w)
if p>=0:
    ans=-math.log10(C)
    print (ans)
else:
    ans=math.log10(C)-math.log10(w)
    print (ans)
if ans < 6.5:
    print('кислая среда')
elif ans > 7.5:
    print('щелочная среда')
else:
    print('нейтральная среда')
```

#### <p style='text-align: justify;'>4. Вычислить средний свободный пробег s (м) молекул в газе при заданных температуре ($\mathrm{^\circ C}$), давлении (атм) и размере молекулы: $s =\frac 1{\pi\sqrt2 d^2 N}$, где $d$ — диаметр молекулы, а $N$ — число молекул в единице объёма. Молекулы рассматриваются как жёсткие сферические частицы. Число молекул в единице объёма можно найти, пользуясь уравнением состояния идеального газа. $d = 3.64\:\mathrm{\mathring{A}}$

Не копируется нормально условие, посмотрите на сервере или напишите как правильно работает синтаксис мне в личку в тг (ник в файле read.me)

контрольный пример
t = 20◦C, P = 1 атм, d = 3.64 ˚A
s = 6.7856e−8 м.

```python
import math
t = float(input('T='))+273.15
P = float(input('P='))*101325
d = float(input('d='))*10**(-10)
#клаперон мендлеев
# pv = nrt
# n/v = p/rt
N = (P*6.02214076e23)/(8.31446261815324*t)
s = 1/(math.pi*(2)**(1/2)*d**(2)*N)
print (s)
```

    #### <p style='text-align: justify;'>5*. Число столкновений молекул $A$ и $B$ (с молекулярными массами $m_A$ и $m_B$, соответственно) в $1\:\mathrm{м^3}$ газа за 1 секунду рассчитывается по формуле: $$z = n_A n_B (r_A + r_B)^2 \sqrt{8\pi RT \left( \frac 1{m_a}+\frac 1{m_B}\right)}$$ Здесь $n_A$ и $n_B$ — число молекул $A$ и $B$ в единице объёма, $r_A$ и $r_B$ — радиусы молекул, $m_A$ и $m_B$ — массы молекул, $T$ — абсолютная температура. Число молекул вида $i$ в единице объёма можно вычислить из уравнения состояния идеального газа. Напишите программу, запрашивающую температуру (в $\mathrm{^\circ C}$) и парциальное давление (в атм) водорода ($r_{\mathrm{H_2}} = 1.365\:\mathrm{\mathring{A}}$) и метана ($r_{\mathrm{CH_4}} = 2.079\:\mathrm{\mathring{A}}$) и выводящую число столкновений молекул.

Не копируется нормально условие, посмотрите на сервере или напишите как правильно работает синтаксис мне в личку в тг (ник в файле read.me)

контрольный пример
T = -0.15◦C,
p(H2) = 0.75 атм,
p(CH4) = 0.25 атм,
z = 9.10434e34 1/(м3·с).

```python
import math
p_H2 = float(input('парциальное давление (в атм) водорода=')) * 101325
p_CH4 = float(input('парциальное давление (в атм) метана=')) * 101325
T = float(input('температура (в цельсиях)=')) + 273.15 

R = 8.3144626181532
Na = 6.022e23 
k = R / Na    #константа больцмана, чтобы было удобнее считать не трогая авагадро
r_H2 = 1.365e-10
r_CH4 = 2.079e-10
m_H2 = 2.016e-3 # фирсов сказал что в контрольном примере брали с точностью до целых эти значения поэтому погрешность
m_CH4 = 16.043e-3
  
n_H2 = p_H2 / (k * T)
n_CH4 = p_CH4 / (k * T)
    
# подставляю всё в формулу по частям, чтобы было проще    
mu = 1 / (1/m_H2 + 1/m_CH4)
cross_section = (r_H2 + r_CH4)**2
relative_velocity = math.sqrt(8 * math.pi * R * T/ mu)
    
z = n_H2 * n_CH4 * cross_section * relative_velocity
    
print(f'Тестовый расчет:')
print(f'z = {z} 1/(м³·с)')
print(f'Ожидаемое значение: 9.10434e34 1/(м³·с)')
print(f'Относительная ошибка: {abs(z - 9.10434e34)/9.10434e34 * 100}%')
```
